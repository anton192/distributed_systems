```python
import numpy as np
import pandas as pd
import mincemeat
from nltk import RegexpTokenizer
from os import listdir
from os.path import join
```

```python
DATA_PATH = './data'
```

```python
file_names = [file_name for file_name in listdir(DATA_PATH)]
```

```python
def map_function(key, value):
    # key -- simple id
    # value = file_name
    
    from nltk import RegexpTokenizer
    from os.path import join
    
    DATA_PATH = './data'
    
    tokenizer = RegexpTokenizer(r'\w+')
    with open(join(DATA_PATH, value), 'r') as document:
        for line in document.readlines():
            for word in tokenizer.tokenize(line):
                yield (value, word.lower()), 1
```

```python
def reduce_function(key, values):
    # key -- <actor>
    # value -- <word>
    return len(values)
```

```python
s = mincemeat.Server()
s.datasource = {i: name for i, name in enumerate(file_names)}
s.mapfn = map_function
s.reducefn = reduce_function

results = s.run_server(port=8095)
```

```python
df = pd.DataFrame(columns=file_names)
df['word'] = np.unique([result[1] for result in results])
df = df.set_index('word')
for (file_name, word), count in results.items():
    df.loc[word][file_name] = count
df
```

```python
df.to_csv('./result.csv')
```
